```python
import pandas as pd
df = pd.read_csv(r"C:\Users\Arun\Downloads\financial_news_events.csv")
df.head()
```

```python
df.drop("News_Url",axis=1,inplace=True)
df.fillna(0,inplace=True)
```

```python
rows=[]
for col in df.columns:
    for i ,row in df.iterrows():
        features=row.drop(col).values
        rows.append({
            "features":features,
            "targetname":col,
            "targetvalue":row[col]
        })
long_df=pd.DataFrame(rows)
```

```python
from sklearn.preprocessing import OneHotEncoder,LabelEncoder
from sklearn.feature_extraction.text import TfidfVectorizer
```

```python
new_encoder=OneHotEncoder(sparse_output=False)
vectorize=TfidfVectorizer()
long_df["features"]=long_df["features"].astype(str)
features=vectorize.fit_transform(long_df["features"])
encoded_target=new_encoder.fit_transform(long_df[["targetname"]])
import numpy as np
x=np.hstack([features.toarray(),encoded_target])
y=long_df["targetvalue"]
```

```python
from sklearn.preprocessing import MinMaxScaler
scaler=MinMaxScaler()
x_new=scaler.fit_transform(x)
```

```python
encoder=LabelEncoder()
y=y.astype(str)
y=encoder.fit_transform(y)
```

```python
print(x.shape)
print(y.shape)
```

```python
y_new=y/499.83
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,random_state=7,test_size=0.3)
new_x_train,new_x_test,new_y_train,new_y_test=train_test_split(x_new,y_new,random_state=7,test_size=0.3)
```

```python
print(x_train.shape)
print(y_train.shape)
```

```python
from catboost import CatBoostRegressor
model=CatBoostRegressor(task_type="GPU")
model.fit(x_train,y_train)
```

```python
new_model=CatBoostRegressor(task_type="GPU")
new_model.fit(new_x_train,new_y_train)
```

```python
y_pred=model.predict(x_test)
new_y_pred=new_model.predict(new_x_test)
```

```python
from sklearn.metrics import r2_score,classification_report
print(r2_score(y_pred,y_test))
print(r2_score(new_y_pred,new_y_test))
```
